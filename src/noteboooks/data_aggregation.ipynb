```python
import numpy as np
import pandas as pd
import json
import os
from textblob import TextBlob
```

```python
path_labeled = "..//data//labeled"
path_all = "..//data//all"
```

```python
def get_dataframe(path):
    df = pd.DataFrame({"question":[],"id":[],"answer":[],"count":[],"cluster":[],"sentiment":[]})
    for _, _, i in os.walk(path):
        for file in i:
            f = json.load(open(f"{path}//{file}", encoding='utf-8'))
            q = f["question"]
            id_ = f["id"]
            for stroke in f["answers"]:
                str_dict = {"question":q, "id":id_}|stroke
                df = df.append(str_dict, ignore_index=True)
    return df
```

```python
df_labeled = get_dataframe(path_labeled)
```

```python
df_all = get_dataframe(path_all)
```

```python
df_all.to_csv("..//data//df_all.csv",encoding="unicode")
```

```python
df_labeled.to_csv("..//data//df_labeled.csv")
```
